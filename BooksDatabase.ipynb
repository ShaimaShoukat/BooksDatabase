```python
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
    except Error as e:
        print(e)
 
    return conn
```

```python
def Choice9(conn):
    """
    9 - Highest rated books enjoyed by each user?
    """
    cur = conn.cursor()
    cur.execute("""select UName,Title,Rating FROM ReadAndRate 
                INNER JOIN User on User.ID=ReadAndRate.UserID
                LEFT JOIN Books on Books.ID=ReadAndRate.BookID
                WHERE Rating>6
                GROUP BY UserID;""")
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)
        
```

```python
def Choice10(conn):
    """
    10 - Find all books that have the same genre as any book written by your favorite author
    """
    FavAuthor = input('Enter your favorite author: ')
    cur = conn.cursor()
    print("\n All books that have the same genre as any book written by"''+FavAuthor+'\n')
    cur.execute("""select DISTINCT(a.Title) from Books a 
                inner join BookCategory b on a.ID=b.BookID
                WHERE CategoryID IN(SELECT CategoryID
                FROM Writes,BookCategory
                WHERE BookCategory.BookID = Writes.BookID
                AND Writes.AuthorID=(select ID from Author where Author=?));""",[FavAuthor])
    rows = cur.fetchall()
 
    for row in rows:
        print(row)
```

```python
def Choice11(conn):
    """
    What Categories can the above user focus on?
    """
    cur = conn.cursor()
    cur.execute("""SELECT DISTINCT(Category.title)
                FROM ReadAndRate 
                INNER JOIN BookCategory on BookCategory.BookID=ReadAndRate.BookID
                LEFT JOIN Category on Category.ID=BookCategory.BookID
                WHERE Rating>7 AND Category.title NOT IN ('Art','Thriller');""")
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)
```

```python
def Choice12(conn):
    """
    Do the books we like all have good ratings?
    """
    cur = conn.cursor()
    cur.execute("""SELECT *
                FROM ReadAndRate
                WHERE BookID IN (SELECT *
                FROM (SELECT BookID
                FROM ReadAndRate
                GROUP BY BookID
                HAVING COUNT(BookID) > 1)   AS a ) ORDER by BookID;""")
                
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)
    print("Yes! The books seem to have similar ratings from dfferent users")
```

```python
def Choice14(conn):
    """
    Authors with most number of highest rated books?
    """
    cur = conn.cursor()
    cur.execute("""select a.Author, count(*) as no_books from Author a 
                inner join Writes b on a.ID=b.AuthorID
                inner join ReadAndRate c on b.BookID=c.BookID
                where Rating>7
                group by b.BookID;""")
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)
```
